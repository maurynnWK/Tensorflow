## 1.Linear Regression
- We are going to use this model to predict survival rate in the titanic dataset

```python
#load data
train_data=pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
test_data=pd.read_csv("https://storage.googleapis.com/tf-datasets/titanic/eval.csv")
y_train=train_data.pop('survived')
y_test=test_data.pop('survived')
```

```python
x_train=train_data#train data here has no label column
x_test=test_data#test data here has no label column
x_train
```

```python
import tensorflow as tf
import tensorflow as tf
categorical_cols=['sex','n_siblings_spouses','parch','class','deck','embark_town','alone']
numerical_cols=['age','fare']

feature_columns=[]
for feature_name in categorical_cols:
  vocabulary=getattr(train_data,feature_name).unique()#list of uniqua values for each column
  feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature_name,vocabulary))

for feature_name in numerical_cols:
    feature_columns.append(tf.feature_column.numeric_column(feature_name,dtype=tf.float32)) 

print(feature_columns)    
```

**INPUT FUNCTION**


```python
def make_input_func(data_df,label_df,num_epochs=10,shuffle=True,batch_size=32):
    def input_function():
        ds=tf.data.Dataset.from_tensor_slices((dict(data_df),label_df))
        if shuffle:
            ds=ds.shuffle(1000)
        ds=ds.batch(batch-size).repeat(num_epochs)
        return ds
    return input_function    
train_input=make_input_func(train_data, y_train)       
test_input=make_input_func(test_data, y_test,num_epochs=1 ,shuffle=False)
```

**MODELLING**

```python
from tensorflow_estimator import estimator
linear_est=estimator.LinearClassifier(feature_columns)

linear_est.train(train_input)
result=linear_est.evaluate(test_input)
print(result['accuracy'])
```
